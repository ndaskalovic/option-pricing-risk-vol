````python
results = np.loadtxt('float32_no_greeks.txt', delimiter=',')
````

````python
X = [int(x) for x in results[:, 0]]
Y = results[:, 1]
````

````python
plt.figure(figsize=(10, 6))
plt.bar(range(len(X)), Y, width=0.6, align='center', edgecolor='black')

plt.xlabel('SIMD vector width')
plt.ylabel('Time (s)')
plt.title('Time to compute 200,000 Monte Carlo paths using SIMD (AVX2, float32)')
plt.xticks(range(len(X)), X)
plt.savefig('simd_speedup_raw.png')
plt.show()
````
